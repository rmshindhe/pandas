#### The is a dataset hosted by the City of New York. The city has an open data platform and they update their information according the amount of data that is brought in. Data were collected through civil birth registration. Each record represents the ranking of a baby name in the order of frequency. Data can be used to represent the popularity of a name.

The task is to seperate the male and female baby names and write them into a excel file with diffrent sheets.

```python
import pandas as pd
URL="https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD"
baby_names = pd.read_csv(URL)
```

```python
baby_names.rename(columns={"Child's First Name":"BN"})
```

```python
baby_names["Child's First Name"].to_frame()
" , ".join(str(name) for name in baby_names["Child's First Name"])
```

```python
baby_names.to_csv("NYC Baby Names.csv", index= False, columns = ['Year of Birth', 'Ethnicity', "Child's First Name", 'Count',
        'Rank'],encoding="utf-8")
```

```python
pd.read_csv("NYC Baby Names.csv")
```

```python
baby_names.columns=['Year of Birth', 'Gender', 'Ethnicity', 'BN', 'Count',
        'Rank']
```

```python
mask=baby_names['Gender']=="FEMALE"
girls=baby_names[mask]
girls
```

```python
mask1=baby_names['Gender'] =="MALE"
boys=baby_names[mask1]
boys
```

```python
baby_names["Gender"]
```

```python
excel_file = pd.ExcelWriter("Baby Names.xlsx")
```

```python
girls.to_excel(excel_file,sheet_name = "Girl", index=False)
boys.to_excel(excel_file,sheet_name = "boy", index = False)
```

```python
excel_file.save()
```
